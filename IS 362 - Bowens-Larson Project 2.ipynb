## Objective
The objective of this project is to practice tidying and analyzing three datasets. 

Dataset 1
The first dataset is list of buildings constructed in 2018 from January to September. The list also contains information about the region the buildings are located in and the number of units in them. 

Dataset 2
The second dataset is a list of weather observations for Central Park from 1869 to 2022.

Dataset 3
The third dataset is a list of test results of five students on two tests from three terms.

## Loading the Data

I started by importing pandas and loading the data that is contained the .CSV files.

```python
# import pandas
import pandas as pd

# load dataset 1 (construction buildings dataset) from .CSV file into a DataFrame
construction_df = pd.read_csv('construction_buildings.csv')

# load dataset 2 (NYC Central Park weather dataset) from .CSV file into a DataFrame
weather_df = pd.read_csv('NYC_Central_Park_weather_1869-2022.csv')

# load dataset 3 (student results dataset) from .CSV file into a DataFrame
student_results_df = pd.read_csv('student_results.csv')

# print DaraFrames
print(construction_df)
print(weather_df)
print(student_results_df)
```

## Tidying the Data

The datasets have mvultiple columns representing different variables which can make it more difficult to analyze them. I reshaped the datasets so that each variable forms a column and each observation forms a row. 

```python
# combine '1 unit', '2 to 4 units', and '5 units or more' into one 'Unit Type' column and save as a new DataFrame
unit = pd.melt(construction_df,
       id_vars=['Year', 'Month'],
       value_vars=['1 unit', '2 to 4 units', '5 units or more'],
       var_name='Unit Type',
              value_name='Unit Count')

# combine 'Northeast', 'Midwest', 'South', and 'West' into one 'Region' column and save as a new DataFrame
region = pd.melt(construction_df,
       id_vars=['Year', 'Month'],
       value_vars=['Northeast', 'Midwest', 'South', 'West'],
       var_name='Region',
                value_name='Region Count')

# combine the two new DataFrames into a new tidy DataFrame
construction_df_tidy = pd.merge(unit, region, on=['Year', 'Month'])

# print the new tidy DataFrame
construction_df_tidy
```

```python
# combine 'PRCP', 'SNOW', 'SNWD', 'TMIN', and 'TMAX' into one 'Weather Variable' column and save as a new DataFrame
weath_df_tidy= pd.melt(weather_df,
       id_vars=['DATE'],
       value_vars=['PRCP', 'SNOW', 'SNWD', 'TMIN', 'TMAX'],
       var_name='WEATHER VARIABLE',
                value_name='VALUE')

# print the new tidy DataFrame
weath_df_tidy
```

```python
# make a copy of the student_results DataFrame
student_results_copy = student_results_df.copy() 

# make 'sex and age' two separate columns
student_results_copy[['sex', 'age']] = student_results_copy['sex and age'].str.split('_', expand=True)

# drop the 'sex and age' column
student_results_copy = student_results_copy.drop(columns=['sex and age'])

# combine the 'term 1', 'term 2', and 'term 3' columns into a 'term' column
student_results_tidy = pd.melt(student_results_copy,
                               id_vars=['id', 'name', 'phone', 'sex', 'age'],
                               value_vars=['term 1', 'term 2', 'term 3'],
                               var_name='term',
                               value_name='score')
# print the new tidy DataFrame
student_results_tidy
```

## Analyzing the Data

### Dataset 1

I analzyed which region has more construction which ould provide insights on factors like resource allocation and market demand. I also analyzed which month has the most and least construction which could aid in making decisions about worforce and investment planning. 

```python
# consctrution per region
construction_per_region = construction_df_tidy.groupby('Region')['Region Count'].sum().sort_values(ascending=True)

print("The region with the most construction: ")
print(construction_per_region[-1:])

# month with most and least construction by region
construction_per_month = construction_df_tidy.groupby('Month')['Region Count'].sum().sort_values(ascending=True)
print("The month with the most construction by region: ")
print(construction_per_month[-1:])

print("The month with the least construction by region: ")
print(construction_per_month[:1])
```

### Dataset 2

I analyed the maximum measurment for each weather occurance. This could help reveal insights into extereme weather events and climate trends.

```python
var_max = weath_df_tidy.groupby('WEATHER VARIABLE')[['DATE','VALUE']].max()
var_max
```

## Analyzing the Data

### Dataset 3

I calculated the average test score and the average score by age. This analysis helps to gain insights on overall performance, effectivness of teaching methods, and developmental trends. 

```python
avg_score = student_results_tidy.score.mean()
print("The average test score was:", avg_score)

avg_score_gender = student_results_tidy.groupby('age').score.mean()
print("The average test score by age was:")
print(avg_score_gender)
```
